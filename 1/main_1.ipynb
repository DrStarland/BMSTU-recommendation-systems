```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# eclat из другой библиотеки, которой нужен иной формат данных
df_eclat = pd.DataFrame(dataset)
df_eclat
```

```python
frequent_itemsets1 = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets1
```

```python
association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.2)
```

```python
frequent_itemsets2 = fpgrowth(df, min_support=0.5, use_colnames=True)
frequent_itemsets2
```

```python
association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.7)
```

```python
# https://github.com/jeffrichardchemistry/pyECLAT/tree/master
from pyECLAT import ECLAT

eclat_instance = ECLAT(data=df_eclat, verbose=False)
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit(min_support=0.25,
                                                           min_combination=1,
                                                           max_combination=3,
                                                           separator=' & ',
                                                           verbose=True)
```

```python
def to_frequent_itemsets(ECLAT_supports):
    freq = {'support': [], 'itemsets': []}
    for key in ECLAT_supports:
        freq['support'].append(get_ECLAT_supports[key])
        freq['itemsets'].append(key.split(' & '))
    freq = pd.DataFrame(freq)
    freq = freq.sort_values(by=['support'], ascending=False)
    return freq

frequent_itemsets3 = to_frequent_itemsets(get_ECLAT_supports)
association_rules(frequent_itemsets3, metric="confidence", min_threshold=0.3)
```

```python
elapsed_time = 0
for i in range(10):
    start_time = time.time()

    # код, время выполнения которого нужно измерить
    for i in range(0, 100):
        # fpgrowth(df, min_support=0.3)
        apriori(df, min_support=0.3)

    # конечное время
    end_time = time.time()
    elapsed_time += (end_time - start_time)

# разница между конечным и начальным временем
print('Затраченное время в пересчёте на один проход алгоритма: ', elapsed_time, 'мс') # / 1000 * 1000
```

```python
graph1 = []
graph2 = []
graph3 = []
axis_x = [j / 100 for j in range(10, 90, 5)]

for sup in axis_x:
    e_time1 = 0
    e_time2 = 0
    e_time3 = 0
    for i in range(10):
        start_time = time.time()
        for i in range(0, 100):
            apriori(df, min_support=sup)
        end_time = time.time()
        e_time1 += (end_time - start_time)

        start_time = time.time()
        for i in range(0, 100):
            eclat_instance.fit(min_support=sup, min_combination=1, max_combination=5, verbose=False)
        end_time = time.time()
        e_time2 += (end_time - start_time)

        start_time = time.time()
        for j in range(0, 100):
            fpgrowth(df, min_support=sup)
        end_time = time.time()
        e_time3 += (end_time - start_time)
    graph1.append(e_time1)
    graph2.append(e_time2)
    graph3.append(e_time3)

# разница между конечным и начальным временем
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.plot(axis_x, graph1, marker='s', color='r', label="Apriori");
plt.plot(axis_x, graph2, marker='s', color='g', label="ECLAT");
plt.plot(axis_x, graph3, marker='s', color='b', label="FP-Growth");
plt.legend()
plt.xlabel('Минимальная поддержка')
plt.ylabel('Время');
```
